```python
# Ploblem- 17

import csv
import random
import datetime

questions_file = "questions.csv"
scores_file = "scores.csv"


def load_questions():
    questions = []
    try:
        with open(questions_file, mode="r", newline="") as file:
            reader = csv.DictReader(file)
            for row in reader:
                questions.append(row)
    except FileNotFoundError:
        with open(questions_file, mode="w", newline="") as file:
            writer = csv.DictWriter(file, fieldnames=["Question ID", "Question Text", "Option A", "Option B", "Option C", "Option D", "Correct Answer", "Marks"])
            writer.writeheader()
    return questions

def save_questions(questions):
    with open(questions_file, mode="w", newline="") as file:
        writer = csv.DictWriter(file, fieldnames=["Question ID", "Question Text", "Option A", "Option B", "Option C", "Option D", "Correct Answer", "Marks"])
        writer.writeheader()
        writer.writerows(questions)


def add_question():
    question_id = input("Enter Question ID: ")
    question_text = input("Enter Question Text: ")
    option_a = input("Enter Option A: ")
    option_b = input("Enter Option B: ")
    option_c = input("Enter Option C: ")
    option_d = input("Enter Option D: ")
    correct_answer = input("Enter Correct Answer (A, B, C, or D): ").upper()
    marks = int(input("Enter Marks for this question: "))
    return {"Question ID": question_id, "Question Text": question_text, "Option A": option_a, "Option B": option_b, "Option C": option_c, "Option D": option_d, "Correct Answer": correct_answer, "Marks": marks}


def update_question(questions):
    question_id = input("Enter the Question ID to update: ")
    for question in questions:
        if question["Question ID"] == question_id:
            question.update(add_question())
            print("Question updated successfully!")
            return
    print("Question not found!")


def delete_question(questions):
    question_id = input("Enter the Question ID to delete: ")
    for question in questions:
        if question["Question ID"] == question_id:
            questions.remove(question)
            print("Question deleted successfully!")
            return
    print("Question not found!")

def conduct_quiz(questions):
    if not questions:
        print("No questions available to conduct the quiz!")
        return

    random.shuffle(questions)
    score = 0
    total_marks = 0
    for question in questions:
        print("\n", question["Question Text"])
        print(f"A: {question['Option A']}  B: {question['Option B']}  C: {question['Option C']}  D: {question['Option D']}")
        answer = input("Enter your answer (A, B, C, or D): ").upper()
        while answer not in ['A', 'B', 'C', 'D']:
            answer = input("Invalid input! Please enter A, B, C, or D: ").upper()
        if answer == question["Correct Answer"]:
            score += int(question["Marks"])
        total_marks += int(question["Marks"])
        print(f"Correct Answer: {question['Correct Answer']}")

    participant_name = input("\nEnter your name: ")
    quiz_date = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    with open(scores_file, mode="a", newline="") as file:
        writer = csv.writer(file)
        writer.writerow([participant_name, quiz_date, score])
    print(f"Quiz Completed! Total Score: {score}/{total_marks}")


def view_scores():
    try:
        with open(scores_file, mode="r") as file:
            print(file.read())
    except FileNotFoundError:
        print("No past scores found!")


def menu():
    questions = load_questions()
    while True:
        print("\n1. Add Question")
        print("2. Update Question")
        print("3. Delete Question")
        print("4. Conduct Quiz")
        print("5. View Scores")
        print("6. Exit")
        choice = input("Enter your choice: ")
        if choice == "1":
            questions.append(add_question())
            save_questions(questions)
        elif choice == "2":
            update_question(questions)
            save_questions(questions)
        elif choice == "3":
            delete_question(questions)
            save_questions(questions)
        elif choice == "4":
            conduct_quiz(questions)
        elif choice == "5":
            view_scores()
        elif choice == "6":
            break
        else:
            print("Invalid choice! Please try again.")

if __name__ == "__main__":
    menu()

```
